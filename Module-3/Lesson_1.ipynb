```python
import matplotlib.pyplot as plt
import numpy as np

pi = 3.1415
```

```python
x = np.arange(0,5*pi,0.1)
```

# Simple Sinusoid

```python
y = 3 * np.sin(x + pi/2)

l1 = np.zeros(int(5*pi/0.1) + 1)
l2 = np.array([pi/2, pi/2])
l3 = np.array([2*pi + pi/2, 2*pi + pi/2])
l4 = np.array([5*pi - pi/2, 5*pi - pi/2])

fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2)

ax.plot(x, l1, 'k-', linewidth=1)
ax.plot(l2, [-5, 5], 'r--', linewidth=1)
ax.plot(l3, [-5, 5], 'r--', linewidth=1)
ax.plot(l4, [-5, 5], 'r--', linewidth=1)

ax.plot(pi/2, 0, 'r.')
ax.plot(2*pi + pi/2, 0, 'r.')
ax.text(pi/2 + 0.2, 0.2, 'A')
ax.text(2*pi + pi/2 + 0.2, 0.2, 'B')

ax.plot(pi, -3, 'r.')
ax.plot(pi, 0, 'r.')
ax.plot(5*pi - pi/2, 0, 'r.')
ax.text(pi + 0.2, -3.2, 'C')
ax.text(pi + 0.2, 0.2, 'D')
ax.text(5*pi - pi/2 + 0.2, 0.2, 'E')

plt.grid()
plt.show()
```

# Signal operations

## Scaling

```python
y1 = np.sin(x + pi/2)
y2 = 3 * y1
l1 = np.zeros(int(5*pi/0.1) + 1)

fig, ax = plt.subplots()

ax.plot(x, y1, '--r', linewidth=2, label="Original: y(t)")
ax.plot(x, y2, linewidth=2, label="Transformed: 3y(t)")
ax.plot(x, l1, 'k-', linewidth=1)

plt.ylim(-5, 5)
plt.grid()
plt.legend()
plt.show()
```

## Changing frequency

```python
y1 = 3 * np.sin(x + pi/2)
y2 = 3 * np.sin(2*x + pi/2)
l1 = np.zeros(int(5*pi/0.1) + 1)

fig, ax = plt.subplots()

ax.plot(x, y1, '--r', linewidth=2, label="Original: y(t)")
ax.plot(x, y2, linewidth=2, label="Transformed: y(2t)")
ax.plot(x, l1, 'k-', linewidth=1)

plt.ylim(-5, 5)
plt.grid()
plt.legend()
plt.show()
```

## Phase Shifting

```python
y1 = 3 * np.sin(x)
y2 = 3 * np.sin(x + pi/2)
l1 = np.zeros(int(5*pi/0.1) + 1)

fig, ax = plt.subplots()

ax.plot(x, y1, '--r', linewidth=2, label="Original: y(t)")
ax.plot(x, y2, linewidth=2, label="Transformed: y(t + π/2)")
ax.plot(x, l1, 'k-', linewidth=1)

plt.ylim(-5, 5)
plt.grid()
plt.legend()
plt.show()
```

## Adding signals

```python
y1 = 3 * np.sin(x)
y2 = 3 * np.sin(2 * x + pi/2)
l1 = np.zeros(int(5*pi/0.1) + 1)

fig, ax = plt.subplots()

ax.plot(x, y1, '--r', linewidth=1, label="$y_1$")
ax.plot(x, y2, ':g', linewidth=1, label="$y_2$")
ax.plot(x, y1 + y2, linewidth=2, label="$y_1$ + $y_2$")
ax.plot(x, l1, 'k-', linewidth=1)

plt.ylim(-7, 7)
plt.grid()
plt.legend()
plt.show()
```

## Multiplying Signals

```python
y1 = 3 * np.sin(x)
y2 = 3 * np.sin(2 * x + pi/2)
l1 = np.zeros(int(5*pi/0.1) + 1)

fig, ax = plt.subplots()

ax.plot(x, y1, '--r', linewidth=1, label="$y_1$")
ax.plot(x, y2, ':g', linewidth=1, label="$y_2$")
ax.plot(x, y1 * y2, linewidth=2, label="$y_1$$y_2$")
ax.plot(x, l1, 'k-', linewidth=1)

plt.ylim(-10, 10)
plt.grid()
plt.legend()
plt.show()
```

## Convolution of signals

```python
y1 = 3 * np.sin(x)
y2 = 3 * np.sin(2 * x + pi/2)
l1 = np.zeros(int(5*pi/0.1) + 1)

fig, ax = plt.subplots()

ax.plot(x, y1, '-r', linewidth=2, label="$y_1$")
ax.plot(x, y2, '-g', linewidth=2, label="$y_2$")
ax.plot(x, np.convolve(y1, y2, 'same'), linewidth=2, label="$y_1$ * $y_2$")
ax.plot(x, l1, 'k-', linewidth=1)

plt.grid()
plt.legend()
plt.show()
```

# Discrete and Continuous Signals

```python
x1 = np.arange(0,5*pi,0.1)
x2 = np.arange(0,5*pi,0.5)
y1 = 3 * np.sin(x1 + pi/2)
y2 = 3 * np.sin(x2 + pi/2)

fig, ax = plt.subplots(2, 1)
ax[0].plot(x1, y1, linewidth=2)
ax[1].plot(x2, y2, '.', linewidth=2)
ax[0].plot(x, l1, 'k-', linewidth=1)
ax[1].plot(x, l1, 'k-', linewidth=1)

ax[0].grid()
ax[1].grid()

plt.tight_layout()
plt.show()
```
